```python
import pandas as pd

# 데이터 불러오기
train_df = pd.read_csv('../data/train.csv')

# train 전체 행 개수
n_rows = len(train_df)

# train에 존재하는 모든 컬럼들의 결측치 개수 및 비율 계산
missing = train_df.isnull().sum()
missing = missing[missing > 0]

missing_percent = (missing / n_rows * 100).round(2)

# 결과 표 만들기
result = pd.DataFrame({
    '결측치 개수': missing,
    '결측치 비율(%)': missing_percent
}).sort_values(by='결측치 비율(%)', ascending=False)

display(result)

# 만약 파일로 저장하고 싶으면 아래 코드 추가
result.to_csv('../data/train_missing_summary.csv', encoding='utf-8-sig')
```
